## 3. Time Series Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

stock='RELIANCE'
```

```python
filled_df.set_index('Date',inplace=True)
```

```python
decomp = seasonal_decompose(filled_df[stock].dropna(),model='multiplicative',period=365)
```

```python
import matplotlib.pyplot as plt

# Plotting the decomposed components
plt.figure(figsize=(12, 8))

# Plot 1: Observed
plt.subplot(411)
plt.plot(decomp.observed, color='blue')
plt.title('Original Data')

# Plot 2: Trend
plt.subplot(412)
plt.plot(decomp.trend, color='green')
plt.title('Trend Component')

# Plot 3: Seasonality
plt.subplot(413)
plt.plot(decomp.seasonal, color='orange')
plt.title('Seasonal Component')

# Plot 4: Residuals
plt.subplot(414)
plt.plot(decomp.resid, color='red')
plt.title('Residual Component')

# Adjust layout and display plot
plt.tight_layout()
plt.show()

# Assigning decomposed components to variables
trend = decomp.trend
seasonal = decomp.seasonal
residual = decomp.resid
```

## 4. Anomaly Detection

```python
filled_df.reset_index(inplace=True)
filled_df
```

```python
ten_stocks_data = filled_df[first_ten_stocks]
ten_stocks_data.set_index(filled_df['Date'], inplace=True)

ten_stocks_data
```

```python
import numpy as np

def detect_anomalies_bollinger_bands(data, window=20, multiplier=2):
    mean_roll = data.rolling(window=window).mean()
    std_roll = data.rolling(window=window).std()

    upper_band = mean_roll + (multiplier * std_roll)
    lower_band = mean_roll - (multiplier * std_roll)

    anomalies = pd.DataFrame(index=data.index)
    anomalies['price'] = data
    anomalies['anomaly'] = (data > upper_band) | (data < lower_band)
    anomalies['deviation'] = np.abs(data - mean_roll)
    return anomalies[anomalies['anomaly']]

anomalies_by_stock_custom = {}
for col in ten_stocks_data.columns:
    anomalies_by_stock_custom[col] = detect_anomalies_bollinger_bands(ten_stocks_data[col])

all_anomalies_custom = pd.concat(anomalies_by_stock_custom, keys=ten_stocks_data.columns, names=['Stock', 'Date'])

top_5_anomalies_custom = all_anomalies_custom.nlargest(5, 'deviation')
top_5_anomalies_df_custom = top_5_anomalies_custom.reset_index()
print(top_5_anomalies_df_custom)
```
